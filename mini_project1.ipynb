```python
import time
start = time.time()
def add_employee(emp_ID,emp_name,emp_DOJ,emp_designation,emp_salary):
    with open('emp_database.txt','a') as e:
        e.write('{}|{}|{}|{}|{}\n'.format(emp_ID,emp_name,emp_DOJ,emp_designation,emp_salary))
    return "\nEmployee added Successfully !!\n"
    

def add_hr(emp_ID,hr_dept,hr_role):
    with open('hr.txt','a') as h:
         h.write('{}|{}|{}\n'.format(emp_ID,hr_dept,hr_role))
    return "\nHR added successfully !!\n"

            
def check(emp_id):
    flag = False
    with open("emp_database.txt", 'r') as emp:
        for i in emp.readlines():
            j = i.split('|')
            if j[0] == emp_id:
                flag = True
    return flag

def search_employee(emp_ID):
    with open('emp_database.txt','r') as e:
        for line in e:
            lst=line.strip().split('|')
            if(lst[0]==emp_ID):
                print("\nEmployee id : {}\nName : {}\nDate of joining : {}\nDesignation : {}\nSalary : {}\n".format(lst[0],lst[1].upper(),lst[2],lst[3],lst[4]))
                break
             

def search_hr(emp_ID):
    with open('hr.txt','r') as hr:
        for line in hr:
            lst=line.strip().split('|')
            if(lst[0]==emp_ID):
                print('\nEmployee id : {}\nHR Department : {}\nHR Role : {}\n'.format(lst[0],lst[1],lst[2]))
                break 
        

while(True):
    print('\n1. Add Record for employee')
    print('2. Add Record for hr')
    print('3. Search a employee Record')
    print('4. Search a hr Record')
    print('5. Exit\n')
    choice = input('Enter your choice: ')
    if choice == '1':
        
        emp_name = input("Enter the name of employee : ")
        emp_DOJ = input("Enter the date of joining : ")
        emp_designation = input("Enter the employee designation : ")
        emp_salary = float(input("Enter the employee salary : "))
        with open('emp_database.txt','r') as e:
            i=0
            for line in e:
                lst=line.strip().split('|')
                i=int(lst[0])
        emp_ID=i+1
        print(add_employee(emp_ID, emp_name, emp_DOJ, emp_designation, emp_salary))
        
    elif choice == '2':
        emp_ID = input("Enter the Employee ID : ")
        if (check(emp_ID)):
            hr_dept = input("HR Department - ")
            hr_role = input("HR Role ")
            print(add_hr(emp_ID, hr_dept, hr_role))
        else:
            print("Employee with this ID does not exist.")

    elif choice == '3':
        emp_id = input('Enter employee code to search: ')
        if(check(emp_id)):
            search_employee(emp_id)
        else:
            print("Employee with this ID does not exist.")
            
    elif choice == '4':
        emp_id = input('Enter the employee code of HR to search: ')
        if(check(emp_id)):
            search_hr(emp_id)
        else:
            print("Employee with this ID does not exist.")
            
    elif choice=='5':
        break
        
    else:
        print('Invalid input')
        
print("time= ",time.time()-start)
```

```python
def check(emp_id):
    flag = False
    with open("emp_database.txt", 'r') as emp:
        for i in emp.readlines():
            j = i.split('|')
            if j[0] == emp_id:
                flag = True
    return flag

def show(emp_id):
    with open('emp_database.txt','r') as e:
        for line in e:
            lst=line.strip().split('|')
            if(lst[0]==emp_id):
                print(lst[3],lst[4]) 
            
def upd_emp(emp_id):
    with open('emp_database.txt','r') as e:
        count=0
        for line in e:
            count+=1
            lst=line.strip().split('|')
            if(lst[0]==emp_id):
                des = input("\nEnter the new designation : ")
                sal = float(input("Enter the new salary : "))
                lst[3] = des
                lst[4] = sal
                print('Updated record : ',lst)
                with open('emp_database.txt','r') as e:
                    i=count-1
                    line = e.readlines()
                    line[i]= ('{}|{}|{}|{}|{}\n'.format(lst[0],lst[1],lst[2],lst[3],lst[4]))
                with open('emp_database.txt','w') as e:
                    e.writelines(line)
                
    
while(True):
    print('\n1.Update employee info.\n2.Exit\n')
    choice = input('Enter your choice: ')
    if choice == '1':
        emp_id = input("Enter employee code to update : ")
        if(check(emp_id)):
            show(emp_id)
            upd_emp(emp_id)
        else:
            print("Employee with this ID does not exist.")
        
    elif choice=='2':
        break
    else:
        print("Invalid input.")
             
```
